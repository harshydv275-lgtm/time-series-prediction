```python
"""
Procurement Prediction - Time Series Forecasting
--------------------------------------------------
Sheets used:
  1. power   -> target column: plant_power__pw_mu
  2. thermal -> target columns: plant_thermal__consumption_mt, plant_thermal__stock_mt

For each target series this script:
  - Loads and cleans the data
  - Splits into train/test (last 30 days as test)
  - Fits an Exponential Smoothing (Holt-Winters) model
  - Evaluates accuracy (MAE, RMSE, MAPE) on the test set
  - Refits on FULL data and forecasts the next N future days
  - Plots Actual vs Test-Forecast vs Future-Forecast
  - Saves forecasts to an Excel file
"""

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

try:
    from statsmodels.tsa.holtwinters import ExponentialSmoothing
    HAVE_STATSMODELS = True
except ImportError:
    HAVE_STATSMODELS = False
    print("NOTE: 'statsmodels' not installed -> using a built-in Holt-Winters "
          "implementation instead. For best results run: pip install statsmodels")

try:
    from sklearn.metrics import mean_absolute_error, mean_squared_error
except ImportError:
    def mean_absolute_error(y_true, y_pred):
        return np.mean(np.abs(np.asarray(y_true) - np.asarray(y_pred)))

    def mean_squared_error(y_true, y_pred):
        return np.mean((np.asarray(y_true) - np.asarray(y_pred)) ** 2)


# --------------------------------------------------------------------------
# FALLBACK: minimal additive Holt-Winters (used only if statsmodels missing)
# --------------------------------------------------------------------------
def simple_holt_winters(series, seasonal_periods, forecast_horizon,
                         alpha=0.3, beta=0.1, gamma=0.3):
    y = series.values.astype(float)
    n = len(y)
    m = seasonal_periods

    level = y[:m].mean()
    trend = (y[m:2 * m].mean() - y[:m].mean()) / m
    seasonals = [y[i] - level for i in range(m)]

    levels, trends, seasonals_hist = [level], [trend], list(seasonals)

    for i in range(n):
        season = seasonals_hist[i % m]
        val = y[i]
        last_level = levels[-1]
        last_trend = trends[-1]

        new_level = alpha * (val - season) + (1 - alpha) * (last_level + last_trend)
        new_trend = beta * (new_level - last_level) + (1 - beta) * last_trend
        new_season = gamma * (val - new_level) + (1 - gamma) * season

        levels.append(new_level)
        trends.append(new_trend)
        seasonals_hist.append(new_season)

    final_level, final_trend = levels[-1], trends[-1]
    last_seasons = seasonals_hist[-m:]

    forecast_vals = [
        final_level + (h + 1) * final_trend + last_seasons[h % m]
        for h in range(forecast_horizon)
    ]
    return np.array(forecast_vals)


class _FallbackModel:
    def __init__(self, series, seasonal_periods):
        self.series = series
        self.seasonal_periods = seasonal_periods

    def forecast(self, steps):
        vals = simple_holt_winters(self.series, self.seasonal_periods, steps)
        future_index = pd.date_range(
            start=self.series.index[-1] + pd.Timedelta(days=1), periods=steps, freq="D"
        )
        return pd.Series(vals, index=future_index)

# --------------------------------------------------------------------------
# CONFIG
# --------------------------------------------------------------------------
FILE_PATH = "time_series__procurement_prediction.xlsx"   # apna path daalo agar alag hai
TEST_DAYS = 30          # last N days held out to check accuracy
FUTURE_DAYS = 30        # kitne din aage forecast karna hai
SEASONAL_PERIODS = 7    # weekly seasonality (daily data)
OUTPUT_FILE = "procurement_forecasts.xlsx"

# --------------------------------------------------------------------------
# HELPER FUNCTION: fit, evaluate, forecast, plot one series
# --------------------------------------------------------------------------
def run_forecast(series: pd.Series, name: str, unit: str):
    series = series.asfreq("D")
    series = series.interpolate()

    # ---- Train/Test split ----
    train = series.iloc[:-TEST_DAYS]
    test = series.iloc[-TEST_DAYS:]

    # ---- Fit on train, evaluate on test ----
    if HAVE_STATSMODELS:
        model_train = ExponentialSmoothing(
            train, trend="add", seasonal="add", seasonal_periods=SEASONAL_PERIODS
        ).fit()
    else:
        model_train = _FallbackModel(train, SEASONAL_PERIODS)

    test_forecast = model_train.forecast(TEST_DAYS)
    test_forecast.index = test.index

    mae = mean_absolute_error(test, test_forecast)
    rmse = np.sqrt(mean_squared_error(test, test_forecast))
    mape = np.mean(np.abs((test - test_forecast) / test)) * 100

    print(f"\n===== {name} =====")
    print(f"MAE  : {mae:.3f} {unit}")
    print(f"RMSE : {rmse:.3f} {unit}")
    print(f"MAPE : {mape:.2f}%")

    # ---- Refit on FULL data, forecast future ----
    if HAVE_STATSMODELS:
        model_full = ExponentialSmoothing(
            series, trend="add", seasonal="add", seasonal_periods=SEASONAL_PERIODS
        ).fit()
    else:
        model_full = _FallbackModel(series, SEASONAL_PERIODS)

    future_forecast = model_full.forecast(FUTURE_DAYS)

    # ---- Plot ----
    plt.figure(figsize=(11, 5))
    plt.plot(series.index, series.values, label="Actual", color="steelblue")
    plt.plot(test_forecast.index, test_forecast.values, label="Test Forecast",
              color="orange", linestyle="--")
    plt.plot(future_forecast.index, future_forecast.values, label="Future Forecast",
              color="red")
    plt.title(f"Exponential Smoothing Forecast - {name}")
    plt.xlabel("Date")
    plt.ylabel(f"{name} ({unit})")
    plt.legend()
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.savefig(f"{name.replace(' ', '_')}_forecast.png", dpi=150)
    plt.show()

    return future_forecast, {"MAE": mae, "RMSE": rmse, "MAPE": mape}


# --------------------------------------------------------------------------
# 1. LOAD DATA
# --------------------------------------------------------------------------
xls = pd.ExcelFile(FILE_PATH)
power = pd.read_excel(xls, sheet_name="power")
thermal = pd.read_excel(xls, sheet_name="thermal")

for d in (power, thermal):
    d["date"] = pd.to_datetime(d["date"])
    d.sort_values("date", inplace=True)
    d.set_index("date", inplace=True)

# --------------------------------------------------------------------------
# 2. RUN FORECASTS FOR ALL TARGET COLUMNS
# --------------------------------------------------------------------------
results = {}
metrics_summary = {}

fc, m = run_forecast(power["plant_power__pw_mu"], "Plant Power Generation", "MU")
results["plant_power__pw_mu"] = fc
metrics_summary["plant_power__pw_mu"] = m

fc, m = run_forecast(thermal["plant_thermal__consumption_mt"], "Thermal Coal Consumption", "MT")
results["plant_thermal__consumption_mt"] = fc
metrics_summary["plant_thermal__consumption_mt"] = m

fc, m = run_forecast(thermal["plant_thermal__stock_mt"], "Thermal Coal Stock", "MT")
results["plant_thermal__stock_mt"] = fc
metrics_summary["plant_thermal__stock_mt"] = m

# --------------------------------------------------------------------------
# 3. SAVE EVERYTHING TO ONE EXCEL FILE
# --------------------------------------------------------------------------
with pd.ExcelWriter(OUTPUT_FILE) as writer:
    for col, fc in results.items():
        fc.to_frame(name=col).to_excel(writer, sheet_name=col[:31])

    metrics_df = pd.DataFrame(metrics_summary).T
    metrics_df.to_excel(writer, sheet_name="accuracy_metrics")

print(f"\nAll forecasts + accuracy metrics saved to: {OUTPUT_FILE}")
```
